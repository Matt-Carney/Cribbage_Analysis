```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
files = [i for i in os.listdir() if 'hand' in i]
dfs = [i.split('.')[0] for i in files]

list_of_dfs = {}
for df, file in zip(dfs,files):
    #print(file)
    plt.figure(figsize=(12,8))
    plt.title('Cribbage ' + file.split('_')[0].capitalize() + ' Card' + ' Hand Value Histogram',
             fontdict = {'fontsize':16})
    list_of_dfs[df] = pd.read_csv(file)
    #list_of_dfs[df] = list_of_dfs[df](columns = {'Percentage of hands':'Percentage of Hands'})
    sns.barplot(list_of_dfs[df]['Score'], list_of_dfs[df]['Percentage of hands'])
    plt.xlabel('Score', fontdict = {'fontsize':12})
    plt.ylabel('Percentage of Hands', fontdict = {'fontsize':12})
    plt.show()
```
